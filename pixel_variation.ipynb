```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from matplotlib.colors import LogNorm
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
```

```python
# Load the FITS files
fits_file1 = "ramp_mode_single_ended/1ramp_0reset_1read/Frame_R0001_M0001_N0001.fits"
fits_file2 = "ramp_mode_single_ended/1ramp_0reset_1read/Frame_R0001_M0001_N0002.fits"

# Open the FITS files and extract the data
with fits.open(fits_file1) as hdul1, fits.open(fits_file2) as hdul2:
    data1 = hdul1[0].data
    data2 = hdul2[0].data

# Compute the difference between the two images
diff = data1 - data2

# Plot the difference with a logarithmic color scale
plt.figure(figsize=(10, 10))
plt.imshow(np.abs(diff), cmap='viridis', norm=LogNorm())
plt.colorbar(label='Pixel Value Difference (log scale)')
plt.title('Difference Between Two Images (Log Scale)')
plt.show()
```

```python
# Mask the differences below the threshold
threshold = 60000
masked_diff = np.ma.masked_less(np.abs(diff), threshold)

# Plot the masked difference
plt.figure(figsize=(10, 10))
plt.imshow(masked_diff, cmap='viridis')
plt.colorbar(label='Pixel Value Difference')
plt.title('Difference Between Two Images (Thresholded)')
plt.show()
```

```python
# Specify the pixel coordinates
i, j = 47, 840

# Plot the time series of pixel values for the specified pixel
plt.figure(figsize=(10, 5))
plt.plot([data1[i, j], data2[i, j]], marker='o', label=f"Pixel ({i}, {j})")
plt.xlabel('Image Number')
plt.ylabel('Pixel Value')
plt.title('Time Series of Pixel Values for Pixel (47, 840)')
plt.legend()
plt.show()
```

```python
# Find the coordinates of the pixels with the largest differences
num_top_pixels = 5  # Number of top pixels to inspect
flat_diff = np.abs(diff).flatten()
top_indices = np.argpartition(flat_diff, -num_top_pixels)[-num_top_pixels:]
top_coords = np.unravel_index(top_indices, diff.shape)

# Print the coordinates and differences of these pixels
for i, j in zip(*top_coords):
    print(f"Pixel ({i}, {j}) has a difference of {diff[i, j]}")

plt.figure(figsize=(10, 5))
for i, j in zip(*top_coords):
    plt.plot([data1[i, j], data2[i, j]], marker='o', label=f"Pixel ({i}, {j})")
plt.xlabel('Image Number')
plt.ylabel('Pixel Value')
plt.title('Time Series of Pixel Values for Top Difference Pixels')
plt.legend()
plt.show()
```

```python
# Coordinates of the pixels with large differences
coords = [(1609, 1261), (1556, 1232), (154, 210), (1556, 1231), (1965, 1775)]

# Print the pixel values in both images for these coordinates
for i, j in coords:
    print(f"Pixel ({i}, {j}) in image 1: {data1[i, j]}")
    print(f"Pixel ({i}, {j}) in image 2: {data2[i, j]}")
    print(f"Difference: {data1[i, j] - data2[i, j]}")
    print()
```

```python
# Open the FITS files and extract the data
fits_file1 = "ramp_mode_single_ended/10ramp_1reset_1read/Frame_R0009_M0001_N0001.fits"
fits_file2 = "ramp_mode_single_ended/10ramp_1reset_1read/Frame_R0010_M0001_N0001.fits"

with fits.open(fits_file1) as hdul1, fits.open(fits_file2) as hdul2:
    data1 = hdul1[0].data.astype(np.int32)  # Convert to a larger data type
    data2 = hdul2[0].data.astype(np.int32)  # Convert to a larger data type

# Compute the difference between the two images
diff = data2 - data1

# Find the coordinates of the pixels with the largest differences
num_top_pixels = 10  # Number of top pixels to inspect
flat_diff = np.abs(diff).flatten()
top_indices = np.argpartition(flat_diff, -num_top_pixels)[-num_top_pixels:]
top_coords = np.unravel_index(top_indices, diff.shape)

for i, j in zip(*top_coords):
    print(f"Pixel ({i}, {j}) has a difference of {diff[i, j]}")

plt.figure(figsize=(10, 5))
for i, j in zip(*top_coords):
    plt.plot([data1[i, j], data2[i, j]], marker='o', label=f"Pixel ({i}, {j})")
plt.xlabel('Image Number')
plt.ylabel('Pixel Value')
plt.title('Time series of pixel values for top difference pixels')
plt.legend()
plt.show()
```

```python
# Coordinates of the pixels with large differences
coords = list(zip(top_coords[0], top_coords[1]))

# Print the pixel values in both images for these coordinates
for i, j in coords:
    print(f"Pixel ({i}, {j}) in image 1: {data1[i, j]}")
    print(f"Pixel ({i}, {j}) in image 2: {data2[i, j]}")
    print(f"Difference: {data1[i, j] - data2[i, j]}")
    print()
```

```python
# Plot the difference with a color bar
plt.figure(figsize=(10, 8))
plt.imshow(diff, cmap='viridis', origin='lower')
plt.colorbar(label='Pixel Value Difference')
plt.title('Difference Between Two Images')
plt.xlabel('X pixels')
plt.ylabel('Y pixels')
plt.show()
```

```python
# Define discrete color levels
levels = MaxNLocator(nbins=15).tick_values(diff.min(), diff.max())
cmap = plt.get_cmap('RdYlGn')
norm = BoundaryNorm(levels, ncolors=cmap.N, clip=True)

# Plot the difference with discrete color levels
plt.figure(figsize=(12, 8))
plt.imshow(diff, cmap=cmap, norm=norm, origin='lower')
plt.xlabel('X pixels')
plt.ylabel('Y pixels')
plt.colorbar(label='Pixel value difference')
plt.title('Difference between 9 & 10 image')
plt.show()
```

```python
coords_less_than_minus_2000 = np.argwhere(diff <= -500)
num_pixels_less_than_minus_2000 = coords_less_than_minus_2000.shape[0]

# Find the coordinates of pixels with a difference of more than 1000
coords_more_than_1000 = np.argwhere(diff > 500)
num_pixels_more_than_1000 = coords_more_than_1000.shape[0]

print(f"Number of pixels with a difference of -500 or less: {num_pixels_less_than_minus_2000}")
print(f"Coordinates of these pixels: {coords_less_than_minus_2000}")

print(f"Number of pixels with a difference of more than 500: {num_pixels_more_than_1000}")
print(f"Coordinates of these pixels: {coords_more_than_1000}")
```
