```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from scipy.ndimage import generic_filter
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
from sklearn.preprocessing import StandardScaler
```

```python
fits_file = "single_ended_single_frame/set_1/im2.fits"
hdul = fits.open(fits_file)
image_data = hdul[0].data.astype(np.int32)
hdul.close()

window_size = 10

def window_std(arr):
    return np.std(arr)

noise_image = generic_filter(image_data, window_std, size=window_size)
```

```python
# Plot the original image
plt.figure(figsize=(10, 8))
plt.imshow(image_data, cmap='gray', origin='lower')
plt.colorbar(label='Pixel Value')
plt.title('Original Image')
plt.xlabel('X Pixels')
plt.ylabel('Y Pixels')
plt.show()
```

```python
# Plot the noise image
plt.figure(figsize=(10, 8))
plt.imshow(noise_image, cmap='viridis', origin='lower')
plt.colorbar(label='Noise (Standard Deviation)')
plt.title('Noise image')
plt.xlabel('X Pixels')
plt.ylabel('Y Pixels')
plt.show()
```

```python
mean_noise = np.mean(noise_image)
std_noise = np.std(noise_image)
min_noise = np.min(noise_image)
max_noise = np.max(noise_image)

print(f"Mean Noise Level: {mean_noise}")
print(f"Standard Deviation of Noise: {std_noise}")
print(f"Minimum Noise Level: {min_noise}")
print(f"Maximum Noise Level: {max_noise}")
```

```python
plt.figure(figsize=(8, 6))
plt.hist(noise_image.flatten(), bins=50, histtype='step', edgecolor='black')
plt.title("Distribution of Spatial Noise")
plt.xlabel("Spatial Noise Level (log scale)")
plt.ylabel("Frequency (log scale)")
plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Flatten noise data for DBSCAN
noise_flat = noise_image.flatten()

# Reshape for clustering (each data point will be a noise value)
X = noise_flat.reshape(-1, 1)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=10)  # You can tweak these parameters
dbscan.fit(X_scaled)

# Get the labels
labels = dbscan.labels_

# Plot the clustered data
plt.figure(figsize=(10, 6))
plt.hist(noise_flat[labels == -1], bins=50, alpha=0.5, label='Noise (outliers)')
for cluster_num in np.unique(labels):
    if cluster_num != -1:
        plt.hist(noise_flat[labels == cluster_num], bins=50, alpha=0.5, label=f'Cluster {cluster_num}')
        
plt.title('Histogram of Spatial Noise by Cluster (DBSCAN)')
plt.xlabel('Spatial Noise Level')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
channel_width = 64  # Assuming 64 columns per channel
n_channels = noise_image.shape[1] // channel_width

mean_noise_per_channel = [
    np.mean(noise_image[:, i*channel_width:(i+1)*channel_width]) 
    for i in range(n_channels)
]

plt.figure(figsize=(10, 6))
plt.plot(range(1, n_channels + 1), mean_noise_per_channel, marker='o', color='teal')
plt.title("Mean Spatial Noise per Channel")
plt.xlabel("Channel Number")
plt.ylabel("Mean Spatial Noise")
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Assuming `cdf_data` contains the CDFs of each channel

# Step 1: Standardize the CDFs to make them comparable
scaler = StandardScaler()
cdf_data_scaled = scaler.fit_transform(np.array(cdf_data))

# Step 2: Apply K-Means Clustering
n_clusters = 2  # Choose the number of clusters (you can tune this based on the data)
kmeans = KMeans(n_clusters=n_clusters, random_state=42, init='k-means++', max_iter=500)
kmeans.fit(cdf_data_scaled)

# Step 3: Assign cluster labels to each channel
labels = kmeans.labels_

# Step 4: Plot the CDFs with clustering results
plt.figure(figsize=(10, 6))

# Different colors for each cluster
colors = plt.cm.get_cmap('tab20', n_clusters)

for i in range(n_clusters):
    # Filter channels belonging to the current cluster
    cluster_channels = np.where(labels == i)[0]
    
    for ch in cluster_channels:
        channel_data = noise_image[:, ch * channel_width:(ch + 1) * channel_width].flatten()
        sorted_data = np.sort(channel_data)
        cdf = np.arange(len(sorted_data)) / float(len(sorted_data) - 1)
        plt.plot(sorted_data, cdf, label=f"Channel {ch+1} (Cluster {i+1})", color=colors(i))

# Title and Labels
plt.title(f"Cumulative Distribution of Spatial Noise per Channel (Clustering with {n_clusters} Clusters)")
plt.xlabel("Spatial Noise Level")
plt.ylabel("Cumulative Probability")
plt.legend(loc="lower right", ncol=2)
plt.grid(True)
plt.show()

# Step 5: Print the cluster centers and sizes
print("Cluster centers (CDF pattern averages):")
print(kmeans.cluster_centers_)

print(f"Cluster Sizes: {np.bincount(labels)}")
```

```python
from sklearn.cluster import DBSCAN

# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(cdf_data_scaled)

# Plot DBSCAN clustering result
plt.figure(figsize=(10, 6))
unique_labels = np.unique(dbscan_labels)
colors = plt.cm.get_cmap('tab20', len(unique_labels))

for label in unique_labels:
    cluster_channels = np.where(dbscan_labels == label)[0]
    for ch in cluster_channels:
        channel_data = noise_image[:, ch * channel_width:(ch + 1) * channel_width].flatten()
        sorted_data = np.sort(channel_data)
        cdf = np.arange(len(sorted_data)) / float(len(sorted_data) - 1)
        plt.plot(sorted_data, cdf, label=f"Channel {ch+1} (Cluster {label})", color=colors(label))

plt.title(f"Cumulative Distribution of Spatial Noise per Channel (DBSCAN Clustering)")
plt.xlabel("Spatial Noise Level")
plt.ylabel("Cumulative Probability")
plt.legend(loc="lower right", ncol=2)
plt.grid(True)
plt.show()
```

```python
# Load the FITS files
filenames = ['single_ended_single_frame/set_1/im{}.fits'.format(i + 1) for i in range(10)]
images = [fits.open(filename)[0].data for filename in filenames]

# Stack the images to create a 3D array
stacked_images = np.stack(images, axis=0)

# Compute the standard deviation across the time dimension (axis=0)
temporal_noise = np.std(stacked_images, axis=0)
```

```python
temporal_noise = np.std(stacked_images, axis=0)

# Plot the temporal noise image
plt.figure(figsize=(10, 8))
plt.imshow(temporal_noise, cmap='viridis', origin='lower')
plt.colorbar(label='Temporal Noise (Standard Deviation)')
plt.title('Temporal noise image')
plt.xlabel('X Pixels')
plt.ylabel('Y Pixels')
plt.show()
```

```python
# Threshold for considering a pixel as "hot"
noise_threshold = 750

# Identify hot pixels
hot_pixels = np.where(temporal_noise > noise_threshold)

# Number of hot pixels
num_hot_pixels = len(hot_pixels[0])

# Average noise per column
average_noise_per_column = np.mean(temporal_noise, axis=0)

print(f'Number of hot pixels: {num_hot_pixels}')
plt.figure(figsize=(10, 6))
plt.plot(average_noise_per_column)
plt.title('Average noise per column')
plt.xlabel('Column Index')
plt.ylabel('Average Noise (Standard Deviation)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Threshold to identify high noise columns
noise_threshold = 30

# Identify columns with average noise above the threshold
high_noise_columns = np.where(average_noise_per_column > noise_threshold)[0]

# Print high noise columns and their average noise values
for col in high_noise_columns:
    print(f"Column {col} has average noise: {average_noise_per_column[col]:.2f}")

# Isolate high noise pixels
high_noise_pixels = np.zeros_like(temporal_noise)
for col in high_noise_columns:
    high_noise_pixels[:, col] = temporal_noise[:, col]

# Plot the high noise pixels
plt.figure(figsize=(10, 6))
plt.imshow(high_noise_pixels, cmap='viridis', origin='lower')
plt.colorbar(label='Noise (Standard Deviation)')
plt.title('High noise pixels')
plt.xlabel('X Pixels')
plt.ylabel('Y Pixels')
plt.show()
```
